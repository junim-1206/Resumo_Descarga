```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('ResumoDescarga_C_82140700.csv', delimiter=';', skiprows=10)
df.reset_index(drop=True, inplace=True)
df['Cota'] = df['Cota'].replace(',', '.', regex=True).astype(float)
df['Vazao'] = df['Vazao'].replace(',', '.', regex=True).astype(float)
# df.drop(41,inplace=True)
df.head()
df.to_excel('Curvas_Limpo.xlsx')
```

```python
sns.lmplot(x="Cota", y="Vazao", data=df, order = 2, ci=None, scatter_kws={"s": 80});
model = np.polyfit(df['Cota'], df['Vazao'], 2)
b2, b1, b0 = model
print(f"y = {b2:.2f}*x^2 + {b1:.2f}*x + {b0:.2f}")
```
